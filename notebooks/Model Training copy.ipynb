Remove Rows with Missing Target

```python
# Remove rows where target is missing
df = df.dropna(subset=['Casualty_class'])

# Check again
df['Casualty_class'].isnull().sum()
```

Handle Missing Feature Values

```python
# Replace all remaining NaN values
df = df.fillna("Unknown")
```

Verify Cleaning

Drop Irrelevant Columns

```python
final_columns = [
    'Time',
    'Day_of_week',
    'Weather_conditions',
    'Road_surface_conditions',
    'Area_accident_occured',
    'Types_of_Junction',
    'Lanes_or_Medians',
    'Casualty_class'
]

# Keep only the final valid features and the target column
existing_columns = [col for col in final_columns if col in df.columns]
df = df[existing_columns]

print("Selected columns:", existing_columns)
print("New shape after selecting final features:", df.shape)
```

Final Check

```python
# Check cleaned dataset
df.head()

print("\nFinal Shape:", df.shape)
print("\nRemaining Columns:\n", df.columns.tolist())
```

Convert Time to Categories

```python
# Convert Time column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')

# Extract hour
df['Hour'] = df['Time'].dt.hour

# Function to map time into categories
def categorize_time(hour):
    if pd.isnull(hour):
        return "Unknown"
    elif 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

# Apply function
df['Time_of_day'] = df['Hour'].apply(categorize_time)

# Drop original Time and Hour (not needed anymore)
df = df.drop(columns=['Time', 'Hour'])

df.head()
```

Create Derived Feature (Day/Night)

```python
# We will not create Day_Night because it is redundant with Time_of_day.
# Keeping only Time_of_day avoids duplicated temporal information.

# No additional derived feature is created here.
print("Using Time_of_day only. Day_Night feature was intentionally omitted.")
```

Simplify Weather Conditions

```python
def simplify_weather(weather):
    if weather in ['Normal']:
        return 'Clear'
    elif weather in ['Raining', 'Rainy', 'Heavy Rain', 'Light Rain']:
        return 'Rain'
    elif weather in ['Fog', 'Foggy', 'Mist']:
        return 'Fog'
    else:
        return 'Other'

df['Weather_conditions'] = df['Weather_conditions'].apply(simplify_weather)

df.head()
```

Simplify Road Surface Conditions

```python
def simplify_road_surface(cond):
    if cond in ['Dry']:
        return 'Dry'
    elif cond in ['Wet or damp', 'Wet']:
        return 'Wet'
    else:
        return 'Other'

df['Road_surface_conditions'] = df['Road_surface_conditions'].apply(simplify_road_surface)

df.head()
```

Create Target Variable (Risk_Level)

```python
def map_risk(casualty):
    if casualty == 'Pedestrian':
        return 'High'
    elif casualty in ['Driver or rider', 'Passenger']:
        return 'Medium'
    else:
        return 'Low'

df['Risk_Level'] = df['Casualty_class'].apply(map_risk)

df.head()
```

Drop Old Target Column

```python
# Drop original casualty column
df = df.drop(columns=['Casualty_class'])

print("Risk_Level distribution:\n", df['Risk_Level'].value_counts())
print("Remaining columns:", df.columns.tolist())
```

Final Check

```python
# Check new columns
df.head()

print("\nUnique Risk Levels:", df['Risk_Level'].unique())
print("\nTime categories:", df['Time_of_day'].unique())
```

Separate Features and Target

```python
# Separate input (X) and target (y)
feature_columns = [
    'Time_of_day',
    'Day_of_week',
    'Weather_conditions',
    'Road_surface_conditions',
    'Area_accident_occured',
    'Types_of_Junction',
    'Lanes_or_Medians'
]

X = df[feature_columns]
y = df['Risk_Level']

print("Selected feature columns:", feature_columns)
print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

Encode Target Variable

```python
# Encode target labels (Low, Medium, High → 0,1,2)
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(df['Risk_Level'])

# Show mapping clearly
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label Mapping:", mapping)
```

One-Hot Encode Input Features

```python
# One-hot encode all categorical features
X_encoded = pd.get_dummies(X)

print("Encoded feature shape:", X_encoded.shape)
X_encoded.head()
```

Feature Scaling

```python
# No scaling is needed for one-hot encoded categorical variables.
# We will use X_encoded directly for model training.

print("No scaling applied to one-hot encoded features.")
```

```python
print("Data is prepared and ready for train/test split.")
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded   # important for class balance
)

class_weights_values = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(zip(np.unique(y_train), class_weights_values))

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Class weights:", class_weights)
```

Convert Target to One-Hot

```python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

print("y_train shape:", y_train_cat.shape)
```

```python
df.head()
```

Build ANN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Define model
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')
])

model.summary()
```

Compile Model

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

Train Model

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

history = model.fit(
    X_train,
    y_train_cat,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    class_weight=class_weights,
    verbose=1
)
```

Plot Training Performance

```python
import matplotlib.pyplot as plt

# Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Evaluate Model (Test Data)

```python
loss, accuracy = model.evaluate(X_test, y_test_cat)

print("Test Accuracy:", accuracy)
print("Test Loss:", loss)
```

Predictions

```python
import numpy as np

# Predict probabilities
y_pred_probs = model.predict(X_test)

# Convert to class index
y_pred = np.argmax(y_pred_probs, axis=1)

# Convert true labels
y_true = np.argmax(y_test_cat, axis=1)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)
print("\nPrediction completed. See classification report below.")
```

Classification Report

```python
from sklearn.metrics import classification_report

class_names = label_encoder.classes_

print("Classification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))
```
